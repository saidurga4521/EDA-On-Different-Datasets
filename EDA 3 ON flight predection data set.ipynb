```python
#import basic liberarys 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_excel('Data_Train.xlsx')
```

```python
df_train.head()
```

```python
df_test=pd.read_excel('Test_set.xlsx')
```

```python
df_test.head()
```

```python
#we can combine the both the test and train data
df=df_train.append(df_test)
```

```python
df.tail()
```

```python
#these all features very difficult so the feature engineering is very difficult.
#feature engineering.
#consider the data of journey feature.
df['date']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['month']=df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['year']=df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df.info()
```

```python
#still these are objects we converted into int.
df['date']=df['date'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
#the next feature is arrival time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_minutes']=df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
df.head(1)
```

```python
# we convert the object into int
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minutes']=df['Arrival_minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#the next feature is the depature time
df['dep_hours']=df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['dep_minutes']=df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#we convert the object into int
df['dep_hours']=df['dep_hours'].astype(int)
df['dep_minutes']=df['dep_minutes'].astype(int)
```

```python
df.info()
```

```python
#next we cosider route and total stops
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()] #then we consider nan as 1
```

```python
#to map the values
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['due_hours']=df['due_hours'].astype(int)
```

```python
#reaminig all are categorey variables so,we do label encoding.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Airline']=le.fit_transform(df['Airline'])
df['Source']=le.fit_transform(df['Source'])
df['Destination']=le.fit_transform(df['Destination'])
df['Additional_Info']=le.fit_transform(df['Additional_Info'])
```
